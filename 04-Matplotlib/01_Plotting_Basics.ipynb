```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, 'co--', label="Virat Kohli")
plt.plot(years, sehwag, 'yd:', label="Virender Sehwag")
    
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
    
plt.grid(True, linestyle='-', linewidth=0.8)
    
plt.show()
```

```python
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

kohli = [1381, 1026, 1268, 1054, 623, 739, 1460, 1202, 1377, 842]

rohit = [611, 168, 1196, 578, 815, 595, 1293, 1030, 1490, 832]

sehwag = [1222, 1173, 860, 684, 540, 420, 300, 210, 150, 90]

plt.plot(years, kohli, 'ro-', label="Kohli")
plt.plot(years, rohit, 'bd-.', label="Rohit")
plt.plot(years, sehwag, 'yo:', label="Sehwag")

plt.xlabel("YEARS")
plt.ylabel("RUNS")
plt.title("COMPARISON")

plt.legend()

plt.tight_layout()
plt.grid(True)
plt.figure(figsize=(10,6))
plt.show()
```

```python
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

kohli = [1381, 1026, 1268, 1054, 623, 739, 1460, 1202, 1377, 842]

rohit = [611, 168, 1196, 578, 815, 595, 1293, 1030, 1490, 832]

sehwag = [1222, 1173, 860, 684, 540, 420, 300, 210, 150, 90]

plt.plot(years, kohli, 'ro-', label="Kohli")
plt.plot(years, rohit, 'bd-.', label="Rohit")
plt.plot(years, sehwag, 'yo:', label="Sehwag")

plt.xticks(years)
plt.xlabel("YEARS")
plt.ylabel("RUNS")
plt.title("COMPARISON")

plt.legend()

plt.tight_layout()
plt.grid(True, linestyle='--', alpha=0.6)
plt.figure(figsize=(10, 6))
plt.show()
```
