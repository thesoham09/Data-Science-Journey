```python
import matplotlib.pyplot as plt
plt.style.use("ggplot") # Choose any style you like
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]


plt.pie(runs, labels=labels)
plt.legend()
plt.title("Career Runs of Indian Batsmen")
```

```python
colors = ['#A8DADC', '#457B9D', '#F1FAEE', '#E63946']
plt.pie(runs, labels=labels, colors=colors)
plt.figure(figsize=(10, 6))

plt.show()
```

```python
#211F1F
plt.pie(
 runs,
 labels=labels,
 colors=colors,
 wedgeprops={'edgecolor': '#211F1F', 'linewidth': 2, 'linestyle': '-'}
)
plt.show()
```

```python
plt.pie(
 runs,
 labels=labels,
 colors=colors,
 wedgeprops={'edgecolor': '#702870', 'linewidth': 2, 'linestyle': '--', 'fill':False}
)
plt.show()
```

```python
explode = [0.1, 0, 0, 0] # Only highlight Sachin's slice
plt.pie(runs, labels=labels, explode=explode, colors=colors)
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode = [0.1, 0, 0, 0] # Only highlight Sachin's slice
plt.pie(runs, labels=labels, explode=explode, colors=colors, shadow=True)
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode = [0.1, 0, 0, 0] # Only highlight Sachin's slice
plt.pie(runs, labels=labels, explode=explode, colors=colors, shadow=True, startangle=200)
plt.title("Exploded Pie Example")
plt.show()
```

```python
explode = [0.1, 0, 0, 0] # Only highlight Sachin's slice
plt.pie(runs, labels=labels, explode=explode, colors=colors, shadow=True, startangle=20, autopct='%1.1f%%')
plt.title("Exploded Pie Example")
plt.show()
```

```python
plt.pie(
 runs,
 labels=labels,
 colors=colors,
 wedgeprops={'edgecolor': '#211F1F', 'linewidth': 2, 'linestyle': '-'}
)
plt.legend()
plt.axis("equal") #EK PROPER CIRCLE MEIN PIE CHART BANTA HAI
plt.show()
```

```python
plt.pie(
 runs,
 labels=labels,
 colors=colors,
 wedgeprops={'edgecolor': '#211F1F', 'linewidth': 2, 'linestyle': '-'}
)
plt.legend()
plt.show()
```
