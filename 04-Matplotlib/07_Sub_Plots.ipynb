```python
import matplotlib.pyplot as plt
```

```python

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']

followers = [1200, 1500, 1800, 2400, 3000, 3800]
likes = [400, 550, 700, 850, 1100, 1500]
reach = [5000, 7000, 8500, 12000, 15000, 18000]
story_views = [300, 450, 600, 750, 950, 1200]

# Create subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Plot 1 → Followers
axs[0, 0].plot(months, followers, color='#7C3AED', marker='o')
axs[0, 0].set_title("Follower Growth")
axs[0, 0].grid(True, alpha=0.3)

# Plot 2 → Likes
axs[0, 1].bar(months, likes, color='#06B6D4')
axs[0, 1].set_title("Monthly Likes")

# Plot 3 → Reach
axs[1, 0].plot(months, reach, color='#EF4444', marker='D')
axs[1, 0].set_title("Reel Reach")
axs[1, 0].grid(True, alpha=0.3)

# Plot 4 → Story Views
axs[1, 1].bar(months, story_views, color='#10B981')
axs[1, 1].set_title("Story Views")

# Main Title
fig.suptitle("Instagram Analytics Dashboard", fontsize=18)

# Spacing
plt.tight_layout()

# Show Plot
plt.show()
```
