```python
import pandas as pd
```

```python
data = {
    'Movie': ['Interstellar', 'Joker', 'Avengers Endgame', 'Parasite'],
    'Director': ['Nolan', 'Todd Phillips', 'Russo Brothers', 'Bong Joon-ho'],
    'Year': [2014, 2019, 2019, 2019],
    'Budget': [165, 70, 356, 11],
    'BoxOffice': [701, 1074, 2798, 262],
    'imdb': [8.6, 8.4, 8.4, 8.5]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.melt(id_vars=["Movie"], value_vars=["Director", "Year", "Budget", "BoxOffice", "imdb"], var_name="Aspects", value_name="Data")
```

```python
df2=df.melt(
    id_vars=["Movie", "Director", "Year"],
    value_vars=["Budget", "BoxOffice", "imdb"],
    var_name="Metric",
    value_name="Value"
)
```

```python
df2
```

```python
#PIVOT
```

```python
df3 = df2.pivot(
    index=["Movie", "Director", "Year"],
    columns="Metric",
    values="Value"
).reset_index()
```

```python
df3
```

```python
#AGGREGATIONS!!
```

```python
df = pd.DataFrame({
    "Dept": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
```

```python
df
```

```python
df.groupby("Dept")["Salary"].mean()
```

```python
df.groupby("Dept")["Salary"].count()
```

```python
df.groupby("Dept")["Salary"].sum()
```

```python
df.groupby("Dept")["Salary"].min()
```

```python
df.groupby("Dept")["Salary"].max()
```

```python
df.groupby("Dept")["Salary"].agg(["mean","max","min"])
```

```python
df.groupby("Dept")["Salary"].agg(AVG="mean",Highest="max",Lowest="min")
```

```python
df.groupby("Team").agg({
"Salary": "mean",
"Age": "max"
})
```

```python
df["AVERAGE"] = df.groupby("Dept")["Salary"].transform("mean")
```

```python
df
```

```python
df.groupby("Team").filter(lambda x: x["Salary"].mean() > 85)
```

```python
df1=df.groupby("Dept").agg({
    "Salary" : "max",
    "Age" : "min"
})
```

```python
df1
```

```python
df1.rename(columns = {"Salary" : "Sal", "Age":"AGE"})
```
